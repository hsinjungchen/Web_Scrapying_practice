```python
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
import json
```

```python
# 啟動瀏覽器工具的選項
options = webdriver.ChromeOptions()

# options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
# options.add_argument("--start-maximized")         #最大化視窗
# options.add_argument("--incognito")               #開啟無痕模式
# options.add_argument("--disable-popup-blocking ") #禁用彈出攔截

# 使用 Chrome 的 WebDriver (含 options)
# driver = webdriver.Chrome(/Users/chenhsinjung/chromedriver)
driver = webdriver.Chrome( options = options )

# 螢幕最大化
#driver.maximize_window()


# 放置爬取的資料
listData = []

# 走訪頁面
def visit():
    driver.get("https://www.104.com.tw/jobs/main/")
    
# 輸入關鍵字
def search():
   

    
    #篩地區
    driver.find_element(By.CSS_SELECTOR, "span#icity").click()
    sleep(2)
    driver.find_element_by_xpath("//input[@value='6001001000']").click()
    sleep(0.5)
    driver.find_element_by_xpath("//input[@value='6001002000']").click()
    sleep(0.5)

    
    btnInput = driver.find_element(By.CSS_SELECTOR, "button.category-picker-btn-primary")
    btnInput.click()
    
    sleep(2)
     # 輸入名稱
    txtInput = driver.find_element(By.CSS_SELECTOR, "input#ikeyword")
    txtInput.send_keys("Python")
    
    
    btnInput2 = driver.find_element(By.CSS_SELECTOR, "button.btn.btn-primary.js-formCheck")
    btnInput2.click()
    
# 滾動頁面
def scroll():
    # 瀏覽器內部的高度
    innerHeightOfWindow = 0
    
    # 當前捲動的量(高度)
    totalOffset = 0
    
    # 在捲動到沒有元素動態產生前，持續捲動
    while totalOffset <= innerHeightOfWindow:
        # 每次移動高度
        totalOffset += 500;
        
        # 捲動的 js code
        js_scroll = '''(
            function (){{
                window.scrollTo({{
                    top:{}, 
                    behavior: 'smooth' 
                }});
            }})();'''.format(totalOffset)
        
        # 執行 js code
        driver.execute_script(js_scroll)
        
        # 強制等待
        sleep(1)
        
        # 透過執行 js 語法來取得捲動後的當前總高度
        innerHeightOfWindow = driver.execute_script('return window.document.documentElement.scrollHeight;');
        
        # 強制等待
        sleep(1)
        
        # 印出捲動距離
        print("innerHeightOfWindow: {}, totalOffset: {}".format(innerHeightOfWindow, totalOffset))
        
        # 為了實驗功能，捲動超過一定的距離，就結束程式
#         if totalOffset >=  5000:
#             break



# 「職缺名稱」、「職缺連結」、「公司名稱」、「公司連結」、「地區」的資訊
def parse():
    python_work = driver.find_elements(By.CSS_SELECTOR, 'div#js-job-content article')

        # 逐一檢視元素
    for element in python_work:
        # 印出分隔文字
        
        
        jName = element.get_attribute("data-job-name")
#         print(jName)
        
        jLink = element.find_element(By.CSS_SELECTOR,'a.js-job-link ')
        jNameLink =jLink.get_attribute('href')
#         print(jNameLink)
        
        cName = element.get_attribute("data-cust-name")
#         print(cName)
        
        cLk = element.find_element(By.CSS_SELECTOR,'li a')
        cLink = cLk.get_attribute('href')
#         print(cLink)
        
        area = element.find_element(By.CSS_SELECTOR,'ul.b-list-inline.b-clearfix.job-list-intro.b-content>li')
        area1 = area.text
#         print(area1)

        listData.append({
        "職缺名稱":jName,
        "職缺連結":jNameLink,
        "公司名稱":cName,
        "公司連結":cLink,
        "地區":area1
        })
    print(listData)


def saveJson():
    fp = open("104jobs.json", "w", encoding='utf-8')    #"w"write:寫入Json
    fp.write( json.dumps(listData, ensure_ascii=False) )
    fp.close()
        
def close():
    driver.quit()
    
# 主程式
if __name__ == '__main__':
    visit()
    search()
    scroll()
    parse()
    saveJson()
    close()
```
